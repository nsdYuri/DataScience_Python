```python
import pandas as pd
notas = pd.read_csv("/ratings.csv") # Atribuindo a base de dadaos a uma variavel
notas.head() # Retorna os 5 primeiros registros do dataframe
```

```python
notas.shape # Retorna a quantidade de linhas e colunas totais da tabela
```

```python
notas.columns = ["usuarioId", "filemId", "nota", "momento"] # Redefinindo o nome das colunas
notas.head()
```

```python
type(notas) # Essa estrutura criada pelo pandas é chamada de dataframe
```

```python
notas['nota'] # Retorna todos os valores da coluna 'notas'
```

```python
notas['nota'].unique() # Informa todos os valores de notas que estão presentes do dataframe
```

```python
notas['nota'].value_counts() # Faz a contagem de cada um dos registros unicos da coluna 'nota' e traz na ordem do mais frequente para menos frequente
```

```python
notas['nota'].mean() # Retona a média de todas as notas presentes na coluna
```
